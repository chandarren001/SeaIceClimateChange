```python
import xarray as xr
import pandas as pd
import netCDF4 as nc
import cartopy
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.path as mpath
import numpy as np
import cftime
from matplotlib.axes import Axes
from cartopy.mpl.geoaxes import GeoAxes
GeoAxes._pcolormesh_patched = Axes.pcolormesh
```

```python
#cell for pr anomaly 2090-2100 vs 2000-2010

#Load data file
#Locate Files
Model_mean = r'C:\Users\Asus\Desktop\Masters Project\Data\pr_actual_9_feb\pr_annual_mean_anomaly.nc' 

#open nc file into xarray dataset
dset_model_mean = xr.open_dataset(Model_mean, decode_times=True, use_cftime=True)

# Create Figure
fig = plt.figure(1, figsize=[10,10])

# Define circle for projection area
theta = np.linspace(0, 2*np.pi, 100)
center, radius = [0.5, 0.5], 0.5
verts = np.vstack([np.sin(theta), np.cos(theta)]).T
circle = mpath.Path(verts*radius+center)

# Discrete colourmap using blue to red colorbar
cmap = mpl.cm.get_cmap('RdBu_r', 21)

# Use cartopy and plot in LambertAzimuthal specify latitude center
ax = plt.subplot(1, 1, 1, projection=ccrs.LambertAzimuthalEqualArea(central_latitude=90))
ax.coastlines()

# Transform projection to plot data onto figure
cs = dset_model_mean['pr'].plot(ax=ax,transform=ccrs.PlateCarree(),vmin = -1.5, vmax = 1.5,cmap = cmap ,add_colorbar = False)

# Set the gridlines and boundaries for projection, crop away outside points with no data
ax.gridlines()
ax.set_boundary(circle, transform=ax.transAxes)
ax.set_extent([-180, 180, 90, 59.5], ccrs.PlateCarree())

# Title name
plt.title('Pr Model Mean Difference between 2000-2010 and 2090-2100 (mm / day)', fontsize=18, y = 1.08)

# Edit colour bar location
cbar_ax = fig.add_axes([0, 0.025, 1.04, 0.05])
cbar = fig.colorbar(cs, cax=cbar_ax, orientation='horizontal',extend='both')

# Colorbar tick size
cbar.ax.tick_params(labelsize=15)

#Save Figure
plt.savefig('C:/Users/Asus/Desktop/Masters Project/Graphs/pr_annual_anomaly_9-feb.png', dpi=300, bbox_inches='tight')
```
